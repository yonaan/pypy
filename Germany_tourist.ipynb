```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow import keras
from tensorflow.keras.optimizers import Adam
```

```python
tb=pd.read_excel(r'Germany_Tourist.xlsx',usecols="B,C",dtype=np.int64)
print(tb)
#dta=tb['Covid Cases'].to_numpy()
no=tb['no'].to_numpy()
arv=tb['Tourist Arrivals'].to_numpy()
print("\n",no,"\n",arv)
```

```python
model=tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
opt = Adam(learning_rate=1e-2, decay=1e-4 / 10)
model.compile(optimizer=opt, loss="mean_squared_error")
xs=np.array(no,dtype=float)
ys=np.array(arv,dtype=float)
model.fit(xs,ys,epochs=1000,batch_size=32)

results=model.predict([51])
print(results)
```
